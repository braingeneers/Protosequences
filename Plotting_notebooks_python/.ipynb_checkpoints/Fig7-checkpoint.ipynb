# Plot Figure 7A-D

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# load datasets
dataPCA = pd.read_excel('load var_pca.xlsx')
data_firing = pd.read_csv('load single_unit_stat_data_org_slice_prim.csv here')
data_red = pd.read_csv('load single_unit_stat_data_org_slice_prim_reduced.csv here')
data_pairs = pd.read_csv('load unit_pair_stat_data_org_slice_prim_reduced.csv here')
# set string identifier for sample type
data_pairs.loc[data_pairs.sample_type == 0, 'sample_type'] = 'mouse_primary'
data_pairs.loc[data_pairs.sample_type == 1, 'sample_type'] = 'organoid'
data_pairs.loc[data_pairs.sample_type == 2, 'sample_type'] = 'mouse_slice'
# edit the scaffold labels
data_firing.loc[data_firing.scaf_labels == 'n', 'scaf_labels'] = 'non-scaffold'
data_firing.loc[data_firing.scaf_labels == 's', 'scaf_labels'] = 'scaffold'
data_red.loc[data_red.scaf_labels == 'n', 'scaf_labels'] = 'non-scaffold'
data_red.loc[data_red.scaf_labels == 's', 'scaf_labels'] = 'scaffold'
# same for the "pairs" dataset
data_pairs.loc[data_pairs.scaf_labels == 'nn', 'scaf_labels'] = 'non-rigid - non-rigid'
data_pairs.loc[data_pairs.scaf_labels == 'ns', 'scaf_labels'] = 'sn'
data_pairs.loc[data_pairs.scaf_labels == 'sn', 'scaf_labels'] = 'scaffold - non-rigid'
data_pairs.loc[data_pairs.scaf_labels == 'ss', 'scaf_labels'] = 'scaffold - scaffold'
# set firing rate to log10
data_firing.firing_rate = np.log10(data_firing.firing_rate)
# make two custom palettes
palette0 = {'scaffold': '#21409a', 'non-scaffold': '#f7941d'}
palette1 = {'non-rigid - non-rigid': '#f7941d',
            'scaffold - non-rigid': '#939598',
            'scaffold - scaffold': '#21409a'}
palette2 = {'organoid': '#66318f', 'mouse_slice': '#2ca9e1', 'mouse_primary': '#2fa148'}
```

```python
# plot firing rate
plt.figure()
fig = sns.violinplot(x='sample_type', y='firing_rate', hue='scaf_labels',
                     data=data_firing, cut=0, scale='area',
                     hue_order=['scaffold', 'non-scaffold'], palette=palette0)
plt.xlabel('')
plt.ylabel('Log10 of firing rate (Hz)')
```

```python
# plot burst to burst correlation
plt.figure()
fig = sns.violinplot(x='sample_type', y='btb_corr', hue='scaf_labels',
                     data=data_red, cut=0, scale='area',
                     hue_order=['scaffold', 'non-scaffold'], palette=palette0)
plt.xlabel('')
plt.ylabel('Normalized burst to burst correlation')
```

```python
# plot pairwise correlations
plt.figure()
fig = sns.violinplot(x='sample_type', y='pw_corr_vals', hue='scaf_labels',
                     data=data_pairs, cut=0, scale='area',
                     hue_order=['scaffold - scaffold', 'scaffold - non-rigid',
                                'non-rigid - non-rigid'], palette=palette1)
plt.xlabel('')
plt.ylabel('Normalized pairwise correlation')
```

```python
order = ['backbone', 'non-rigid', 'all']
# plot normalized PCa variance
plt.figure()
fig = sns.violinplot(x='unit_type', y='value', density_norm='width', cut=0,
                     hue='sample_type', inner='point', palette=palette2,
                     order=order, data=dataPCA)
plt.ylabel('Normalized PCA variance')
```
