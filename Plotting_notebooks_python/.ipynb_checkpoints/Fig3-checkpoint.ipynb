# Plot Figure 3E

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# load data_firing
data = pd.read_excel('Q:/Personal/Mattia/OrganoidsFigures/figures/figure3_files/corr_pairs.xlsx')
# folder 2 save figures
folder2save = 'Q:/Personal/Mattia/OrganoidsFigures/figures/figure3_files/'
```

```python
data.columns
```

```python
data['pos'] = data.organoid*10 + data.pairID
# plot peak_times
plt.figure()
fig = sns.violinplot(x='pos', y='corr',
                     density_norm='width', cut=0,
                     data=data)
plt.xlabel('')
plt.ylabel('')
# save figure
plt.savefig(folder2save + 'pair_corr.svg')
```
