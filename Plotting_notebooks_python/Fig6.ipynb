# Plot Figure 6F-G

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# load datasets
datab2b = pd.read_excel('load b2bcorr.xlsx here')
data_corr = pd.read_excel('load corr_pairs.xlsx here')
```

```python
# make a palette for the plot
my_pal = {1: '#21409a', 2: '#f7941d', 4: '#21409a', 5: '#f7941d',
          7: '#21409a', 8: '#f7941d', 10: '#21409a', 11: '#f7941d',
          10: '#21409a', 11: '#f7941d', 13: '#21409a', 14: '#f7941d',
          16: '#21409a', 17: '#f7941d'}
# plot burst to burst correlation; blue are backbone units, yellow non-rigid ones
plt.figure()
fig = sns.violinplot(x='group', y='b2bcorr', cut=0,
                     density_norm='width', palette=my_pal,
                     data=datab2b)
plt.xlabel('Individual cortical mouse slice')
plt.ylabel('Burst to burst correlation')
```

```python
# create a unique identifier for "unit pair type" and organoid
data_corr['pos'] = data_corr.organoid*10 + data_corr.pairID
# make a palette for the plot
my_pal = {11: '#21409a', 12: '#939598', 13: '#f7941d',
          21: '#21409a', 22: '#939598', 23: '#f7941d',
          31: '#21409a', 32: '#939598', 33: '#f7941d',
          41: '#21409a', 42: '#939598', 43: '#f7941d',
          51: '#21409a', 52: '#939598', 53: '#f7941d',
          61: '#21409a', 62: '#939598', 63: '#f7941d'}
# plot peak_times
plt.figure()
fig = sns.violinplot(x='pos', y='corr', density_norm='width',
                     cut=0, palette=my_pal, data=data_corr)
plt.xlabel('Individual cortical mouse slice')
plt.ylabel('Pairwise cross-correlation')
```
