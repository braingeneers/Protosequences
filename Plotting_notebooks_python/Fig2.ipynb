# Plot Figure 2E-F

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# load data_firing
data = pd.read_excel('load peak_times.xlsx here')
datab2b = pd.read_excel('load b2bcorr.xlsx here')
```

```python
# make a palette for the plot
my_pal = {1: '#21409a', 2: '#f7941d', 4: '#21409a', 5: '#f7941d',
          7: '#21409a', 8: '#f7941d', 10: '#21409a', 11: '#f7941d'}
# plot burst to burst correlation; blue are backbone units, yellow are non-rigid units
plt.figure()
sns.violinplot(x='group', y='b2bcorr',
               scale='width', cut=0,
               palette=my_pal, data=datab2b)
plt.xlabel('Individual organoid')
plt.ylabel('Burst to burst correlation')
```

```python
# plot the peak time for the example organoid - density plot and scatter plot on top
plt.figure()
sns.kdeplot(x='unit', y='peak_time', fill=True,
            common_norm=False, data=data, cut=0,
            cmap='Reds')
sns.scatterplot(x='unit', y='peak_time',
                color='k', data=data)
plt.xlabel('Unit ID (ordered)')
plt.ylabel('Peak time from burst peak (ms)')
plt.ylim(-200, 600)
# remove the label
plt.legend([],[], frameon=False)
```
