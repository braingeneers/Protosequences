# Plot Figure 3E

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# load data_firing
data = pd.read_excel('load corr_pairs.xlsx here')
```

```python
# create a unique identifier for "unit pair type" and organoid
data['pos'] = data.organoid*10 + data.pairID
# make a palette for the plot
my_pal = {11: '#21409a', 12: '#939598', 13: '#f7941d',
          21: '#21409a', 22: '#939598', 23: '#f7941d',
          31: '#21409a', 32: '#939598', 33: '#f7941d',
          41: '#21409a', 42: '#939598', 43: '#f7941d'}
# plot burst to burst correlation
# blue are backbone-backbone unit pairs,
# gray are backbone-non-rigid unit pairs,
# yellow are non-rigid-non-rigid unit pairs,

plt.figure()
fig = sns.violinplot(x='pos', y='corr', density_norm='width',
                     cut=0, palette=my_pal, data=data)
plt.xlabel('Individual organoid')
plt.ylabel('Pairwise cross-correlation')
```
